```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn import metrics
from imblearn.over_sampling import SMOTE
import pickle
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix

# Step 1: Replace 'NA' string values with actual NaN values (if 'NA' is present in the dataset)
data.replace('NA', np.nan, inplace=True)

# Step 2: Handle Missing Values (NaN)
cat_imputer = SimpleImputer(strategy='most_frequent')  # Categorical imputer
num_imputer = SimpleImputer(strategy='mean')  # Numerical imputer

# Impute missing values for categorical columns
data[['Race ', 'Age', 'genotype']] = cat_imputer.fit_transform(
    data[['Race ', 'Age', 'genotype']]
)

# Impute missing values for numerical columns
data[['Height', 'Weight ', 'INR on ReportedDose', 'Therapeutic Dow']] = num_imputer.fit_transform(
    data[['Height', 'Weight ', 'INR on ReportedDose', 'Therapeutic Dow']]
)

# Step 3: Encode Categorical Variables
label_encoder = LabelEncoder()
for col in ['Race ', 'Age', 'genotype']:
    data[col] = label_encoder.fit_transform(data[col])

# Ensure the 'Gender' column is present before dropping
if 'Gender' in data.columns:
    data = data.drop(columns=['Gender'])
else:
    print("'Gender' column not found!")

# Step 4: Split data into features (X) and target (Y)
X = data.drop('Therapeutic Dow', axis=1)
Y = data['Therapeutic Dow']

# Step 5: Check class distribution
print("Class distribution in target variable Y before splitting:")
print(Y.value_counts())

# Handle potential issues with a single sample in any class
if Y.value_counts().min() == 1:
    print("Warning: One of the classes has only one sample. This will be handled.")
    classes_with_one_sample = Y.value_counts()[Y.value_counts() == 1].index
    data = data[~data['Therapeutic Dow'].isin(classes_with_one_sample)]

# Step 6: Use stratified sampling to split dataset into training and test sets
X = data.drop('Therapeutic Dow', axis=1)
Y = data['Therapeutic Dow']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)

# Step 7: Convert continuous target variable to categorical labels using quantile-based binning
Y_train_cat, bins = pd.qcut(Y_train, q=3, labels=False, retbins=True, duplicates='drop')
num_bins = len(bins) - 1  # Determine number of bins
labels = list(range(num_bins))  # Create dynamic labels
Y_train_cat = pd.qcut(Y_train, q=num_bins, labels=labels, retbins=False, duplicates='drop')
Y_test_cat = pd.cut(Y_test, bins=bins, labels=labels)

# Handle NaN in target categories
valid_indices = ~pd.isnull(Y_test_cat)
Y_test_cat = Y_test_cat[valid_indices]
X_test = X_test[valid_indices]

# Check if there is more than one class in Y_train_cat
if len(np.unique(Y_train_cat)) > 1:
    # Step 8: Apply SMOTE for class imbalance on the binned target variable
    print("Class distribution before SMOTE:")
    print(pd.Series(Y_train_cat).value_counts())

    smote = SMOTE(random_state=42)
    X_train_smote, Y_train_smote = smote.fit_resample(X_train, Y_train_cat)

    print("\nClass distribution after SMOTE:")
    print(pd.Series(Y_train_smote).value_counts())

    # Step 9: Visualize class distribution after SMOTE
    plt.bar(pd.Series(Y_train_smote).value_counts().index.astype(str),
            pd.Series(Y_train_smote).value_counts().values)
    plt.title("Class Distribution After SMOTE")
    plt.xlabel("Class")
    plt.ylabel("Frequency")
    plt.show()
else:
    print("Only one class found in the target variable. SMOTE will not be applied.")

# Step 10: Define parameter grid for GridSearchCV
param_grid = {
    'max_depth': [5, 10, 15, 20],
    'min_samples_leaf': [2, 4, 6, 8],
}

# Initialize the DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(random_state=42)

# Perform GridSearchCV with cross-validation
grid_search = GridSearchCV(
    estimator=dt_classifier,
    param_grid=param_grid,
    cv=5,  # 5-fold cross-validation
    scoring='accuracy',  # Use accuracy for classification
    verbose=1,
    n_jobs=-1  # Use all available cores
)
grid_search.fit(X_train_smote, Y_train_smote)

# Step 11: Retrieve the best model from GridSearchCV
best_dt_model = grid_search.best_estimator_

# Step 12: Make predictions on the training and test sets
y_train_pred = best_dt_model.predict(X_train_smote)
y_test_pred = best_dt_model.predict(X_test)

# Step 13: Evaluate the model
train_accuracy_dt = metrics.accuracy_score(Y_train_smote, y_train_pred)
test_accuracy_dt = metrics.accuracy_score(Y_test_cat, y_test_pred)
train_mse_dt = metrics.mean_squared_error(Y_train_smote, y_train_pred)

print(f"Train Accuracy: {train_accuracy_dt}")
print(f"Test Accuracy: {test_accuracy_dt}")
print(f"Train Mean Squared Error: {train_mse_dt}")

# Best model parameters and score from GridSearchCV
print("Best Parameters: ", grid_search.best_params_)
print("Best Grid Search Score: ", grid_search.best_score_)

# Step 14: Save the best model for later use
with open('best_decision_tree_model.pkl', 'wb') as file:
    pickle.dump(best_dt_model, file)

# Step 15: Perform additional cross-validation for evaluation
cross_val_scores_dt = cross_val_score(best_dt_model, X_train_smote, Y_train_smote, cv=5, scoring='accuracy')

print(f"Cross-Validation Accuracy Scores: {cross_val_scores_dt}")
print(f"Mean Cross-Validation Accuracy: {cross_val_scores_dt.mean()}")

# Step 16: Generate classification report and confusion matrix
print("Classification Report (Test Set):")
print(classification_report(
    Y_test_cat,
    y_test_pred,
    zero_division=0  # Suppresses warnings by assigning 0 to undefined precision/recall
))

print("\nConfusion Matrix (Test Set):")
print(confusion_matrix(Y_test_cat, y_test_pred))
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import xgboost as xgb
from imblearn.over_sampling import ADASYN
import matplotlib.pyplot as plt

# Step 1: Data Preprocessing
# Assuming `data` is your dataset
data.replace('NA', np.nan, inplace=True)

# Step 2: Handle Missing Values (NaN)
cat_imputer = SimpleImputer(strategy='most_frequent')  # Categorical imputer
num_imputer = SimpleImputer(strategy='mean')  # Numerical imputer

# Impute missing values for categorical columns
data[['Race ', 'Age', 'genotype']] = cat_imputer.fit_transform(
    data[['Race ', 'Age', 'genotype']]
)

# Impute missing values for numerical columns
data[['Height', 'Weight ', 'INR on ReportedDose', 'Therapeutic Dow']] = num_imputer.fit_transform(
    data[['Height', 'Weight ', 'INR on ReportedDose', 'Therapeutic Dow']]
)

# Step 3: Encode Categorical Variables
label_encoder = LabelEncoder()
for col in ['Race ', 'Age', 'genotype']:
    data[col] = label_encoder.fit_transform(data[col])

# Ensure the 'Gender' column is present before dropping
if 'Gender' in data.columns:
    data = data.drop(columns=['Gender'])
else:
    print("'Gender' column not found!")

# Step 4: Discretize the Target Variable (if it's continuous)
# Assuming 'Therapeutic Dow' is continuous, let's bin it into 3 categories for classification
data['Therapeutic Dow'] = pd.cut(data['Therapeutic Dow'], bins=3, labels=[0, 1, 2])

# Step 5: Split data into features (X) and target (Y)
X = data.drop('Therapeutic Dow', axis=1)
Y = data['Therapeutic Dow']

# Step 6: Use stratified sampling to split dataset into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)

# Step 7: Handle class imbalance with ADASYN
adasyn = ADASYN(random_state=42)
X_train_balanced, Y_train_balanced = adasyn.fit_resample(X_train, Y_train)

# Step 8: Hyperparameter optimization for XGBoost using RandomizedSearchCV
param_dist = {
    'n_estimators': [100, 200, 300, 400, 500],
    'learning_rate': [0.001, 0.01, 0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7, 9],
    'subsample': [0.5, 0.6, 0.7, 0.8, 1.0],
    'colsample_bytree': [0.5, 0.6, 0.7, 0.8, 1.0],
    'reg_alpha': [0, 0.1, 0.5, 1.0],
    'reg_lambda': [0, 0.1, 1.0, 5.0]
}

# Initialize the XGBClassifier without the use_label_encoder parameter
xgb_model = xgb.XGBClassifier(random_state=42)

random_search = RandomizedSearchCV(xgb_model, param_distributions=param_dist, n_iter=50, cv=5, verbose=1, random_state=42, n_jobs=-1)
random_search.fit(X_train_balanced, Y_train_balanced)

# Best model and hyperparameters after RandomizedSearchCV
best_xgb_model = random_search.best_estimator_

# Step 9: Make predictions
y_train_pred = best_xgb_model.predict(X_train_balanced)
y_test_pred = best_xgb_model.predict(X_test)

# Step 10: Evaluate the model
train_accuracy_xgb = accuracy_score(Y_train_balanced, y_train_pred)
test_accuracy_xgb = accuracy_score(Y_test, y_test_pred)

print(f"Train Accuracy (XGBoost): {train_accuracy_xgb}")
print(f"Test Accuracy (XGBoost): {test_accuracy_xgb}")

# Classification report and confusion matrix
print("Classification Report (Test Set):")
print(classification_report(Y_test, y_test_pred, zero_division=0))

print("\nConfusion Matrix (Test Set):")
print(confusion_matrix(Y_test, y_test_pred))

# Step 11: Cross-validation for model evaluation
cross_val_scores = cross_val_score(best_xgb_model, X_train_balanced, Y_train_balanced, cv=5, scoring='accuracy')

print(f"Cross-Validation Accuracy Scores (XGBoost): {cross_val_scores}")
print(f"Mean Cross-Validation Accuracy (XGBoost): {cross_val_scores.mean()}")

# Step 12: Save the best model
import pickle
with open('best_xgb_model.pkl', 'wb') as file:
    pickle.dump(best_xgb_model, file)

# Visualizing class distribution after ADASYN
plt.bar(pd.Series(Y_train_balanced).value_counts().index.astype(str),
        pd.Series(Y_train_balanced).value_counts().values)
plt.title("Class Distribution After ADASYN")
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import silhouette_score
import pickle

# Assuming X_train_scaled is your scaled dataset and contains the features
# Assuming Y_train contains the actual therapeutic dose values
n_clusters = 5  # Number of bins/clusters for therapeutic dosage

# Step 1: Apply K-Means Clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42)  # Set the number of clusters
kmeans.fit(X_train_scaled)

# Get cluster labels for the training set
y_kmeans = kmeans.predict(X_train_scaled)

# Step 2: Evaluate the clustering with Silhouette Score
sil_score_kmeans = silhouette_score(X_train_scaled, y_kmeans)
print(f"Silhouette Score (K-Means): {sil_score_kmeans}")

# Step 3: Create a DataFrame for analysis
df_clustered = pd.DataFrame(X_train_scaled, columns=[
    "Gender", "Race", "Age", "Height", "Weight",
    "Diabetes", "Simvastatin", "Amiodarone", "INR"
])
df_clustered["Cluster"] = y_kmeans

# Step 4: Map therapeutic dosage bins to clusters using pd.cut
# Use pd.cut instead of pd.qcut
therapeutic_dose_labels = ['0', '1', '2', '3', '4']  # Labels for therapeutic dose bins
bins = pd.cut(Y_train, bins=n_clusters, labels=therapeutic_dose_labels, retbins=True)
df_clustered["Therapeutic Dose Class"] = bins[0]  # Assign bin labels to the DataFrame
bin_edges = bins[1]  # Get bin edges

# Print the ranges of each therapeutic dose class
print("\nRanges for each Therapeutic Dose Class:")
for i in range(len(bin_edges) - 1):
    print(f"Class {therapeutic_dose_labels[i]}: {bin_edges[i]:.2f} - {bin_edges[i+1]:.2f}")

# Step 5: Print the clustered DataFrame with therapeutic dose classes
print("\nClustered DataFrame with Therapeutic Dose Classes:")
print(df_clustered.head())

# Step 6: Visualize the clusters
plt.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X')  # Centroids
plt.title('K-Means Clustering based on Therapeutic Dose')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster')
plt.show()

# Step 7: Save the K-Means model
with open('kmeans_model_therapeutic_dose.pkl', 'wb') as file:
    pickle.dump(kmeans, file)

# Step 8: Print cluster definitions with therapeutic dose ranges
print("\nCluster Definitions based on Therapeutic Dose:")
for cluster in range(n_clusters):
    print(f"Cluster {cluster}: Therapeutic Dose Class {therapeutic_dose_labels[cluster]}")
```
